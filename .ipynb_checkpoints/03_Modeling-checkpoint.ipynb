```python
#function to run chosen model
def model_predictions(x, X_train, X_test, y_train, y_test):
    #global X_train, X_test, y_train, y_test, model, predictions
    if x == "logistic":
        model = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial')
        model.fit(X_train, y_train)
        print("Logistic:")
        predictions =  model.predict(X_test)
        labels = list(y_train.unique()).sort()
    elif x == "knn":
        weights = ['uniform', 'distance']
        n_neighbors=5
        model = KNeighborsClassifier(n_neighbors, weights = weights[1])
        model.fit(X_train, y_train)
        print("KNN:")
        predictions = model.predict(X_test)
        labels = list(mapping.keys())
    elif x=="decision_tree":
        model = DecisionTreeClassifier()
        model.fit(X_train, y_train)
        print("Decision Tree:")
        predictions = model.predict(X_test)
        labels = list(mapping.keys())
    elif x=="random_forest":
        model = RandomForestClassifier()
        model.fit(X_train, y_train)
        print ("Random Forest:")
        predictions = model.predict(X_test)
        labels = list(mapping.keys())
    else:
        print ("Input should be logistic, knn, decision_tree or random_forest")   
    
    #calculate scores. Kappa/ROC-AUC/F1 were chosen and should be as high as possible 0-1
    print("The accuracy of the model on test set is: ", accuracy_score(y_test, predictions)) 
    print("Kappa: ", cohen_kappa_score(y_test, predictions))  
    #print("Avg. precision: ", average_precision_score(y_test, predictions))
    #print("ROC-AUC score: ", roc_auc_score(y_test, predictions)) 
    print("F1 score: ", f1_score(y_test, predictions, average = 'macro'))  
    #generate confusion matrices with raw and normalised values
    print ("Confusion Matrix - raw:")
    a4_dims = (32, 12)
    fig, ax = plt.subplots(figsize=a4_dims)
    plot_confusion_matrix(model, X_test, y_test, cmap = "PuBuGn", ax=ax, labels = labels)
    plt.grid(False)
    plt.show() 
    fig2, ax2 = plt.subplots(figsize=a4_dims)
    print ("Confusion Matrix - normalised row-wise:")
    plot_confusion_matrix(model, X_test, y_test, normalize='true', cmap = "PuBuGn", ax=ax2, 
                          labels = labels)
    plt.grid(False) 
    plt.show()
    return [model, predictions]
```

```python
def oversampler_choice_train(x):
    global X_train, y_train
    if x=="ros":
        ros = RandomOverSampler()
        X_train, y_train = ros.fit_sample(X_train, y_train)
        print("Random Oversampler:\n", y_train.value_counts())
    elif x=="smote":
        print(y_train.value_counts())
        smote = SMOTE()
        X_train, y_train = smote.fit_sample(X_train, y_train)
        print("SMOTE:\n", y_train.value_counts())
    elif x == "ada":
        ada = ADASYN(random_state=42)
        X_train, y_train = ada.fit_sample(X_train, y_train)
        print("ADASYN:\n", y_train.value_counts())
    else:
        print ("Input should be ros, smote or ada")
```

```python
n_clust = 16
def get_kmeans_ineratia_varying_max_iter(max_iter):
    kmeans = KMeans(n_clusters=n_clust,
                    random_state=1234,
                    n_init=3,
                    algorithm='elkan',
                    max_iter=max_iter,
                   )
    kmeans.fit(X)

    return kmeans.inertia_

max_iter_list = [1, 5, 10, 20, 30, 40, 50, 100,500,1000]

plt.plot(max_iter_list,
         [get_kmeans_ineratia_varying_max_iter(x) for x in max_iter_list],
        )
plt.xlabel('Max iter')
plt.ylabel('inertia')
```

```python
kmeans = KMeans(n_clusters=16, random_state=1234, max_iter=200)
kmeans.fit(X)

# print the cluster centers
clusters_coordinates = kmeans.cluster_centers_
clusters = kmeans.predict(X)
```

```python
X_UL = data.copy()
X_UL['cluster'] = clusters
```

# lets see how the cluster vary among the genres:

```python
hue_order = list(X_UL['main_genre'].unique())
hue_order.sort()
```

```python
palette = sns.color_palette("gist_rainbow", 16)
a4_dims = (16, 7)
for n in range(4):
    fig, ax = plt.subplots(figsize=a4_dims)
    b = sns.countplot(x="cluster", hue="main_genre", 
                  data=X_UL[(X_UL["cluster"] >= 4*(n)) & (X_UL["cluster"] < 4*(n+1))], 
                  ax=ax, palette = palette, hue_order=hue_order)
    b.set_xlabel("cluster",fontsize=16)
    b.set_ylabel("count",fontsize=16)
    b.tick_params(labelsize=14)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.setp(ax.get_legend().get_texts(), fontsize='18') # for legend text
    plt.setp(ax.get_legend().get_title(), fontsize='18') # for legend title
    plt.show()
```

```python
K = range(2,50)
inertias = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234,
                   )
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)
    
plt.figure(figsize=(16,8))
plt.plot(K, inertias, 'b', linewidth=4.0)
plt.xlabel('Clusters', fontsize=14)
plt.ylabel('inertia', fontsize=14)
plt.xticks(np.arange(min(K), max(K)+1, 2), fontsize=14)
plt.yticks(fontsize=14)
plt.title(f'Elbow Method showing the optimal number of clusters', fontsize=16,fontweight='bold')
```

```python
K = range(2,50)
silhouettes = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234,
                   )
    kmeans.fit(X)
    silhouettes.append(silhouette_score(X,
                                     kmeans.predict(X)
                                    )
                   )

plt.figure(figsize=(16,8))
plt.plot(K, silhouettes, 'b', linewidth=4.0)
plt.xlabel('clusters', fontsize=14)
plt.ylabel('silhouette score', fontsize=14)
plt.xticks(np.arange(min(K), max(K)+1, 2), fontsize=14)
plt.yticks(fontsize=14)
plt.title('Elbow Method showing the optimal number of clusters', fontsize=16,fontweight='bold')
```
